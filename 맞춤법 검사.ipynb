```python
from hanspell import spell_checker

text = "아버지가방에들어가신다나는오늘코딩을했다"

spell_checker.check(text).as_dict()['checked']
```

```python
ind = set(df[df['title'].str.contains('워')].index) - set(df[df['title'].str.contains('워라')].index)
df.loc[list(ind)]
```

```python
df['title'] = df['title'].apply(lambda x: x.replace('워라벨', '워라밸'))
df['title']
```

```python
df['title'] = df['title'].apply(lambda x: x.replace('“', ''))
df['title'] = df['title'].apply(lambda x: x.replace('”', ''))
df['title']
```

```python
df[df['title'].str.contains('와 라')]
```

```python
spelling = []
for text in df['title']:
    spelling.append(spell_checker.check(text).as_dict()['checked'])
```

```python
spelling
```

```python
df['title'] = spelling
df['title']
```

```python
df.to_csv('blind_sk_spell.csv')
df.head()
```
