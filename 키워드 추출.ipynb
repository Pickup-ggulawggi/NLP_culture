```python
import pandas as pd
import re
```

```python
df = pd.read_csv('blind_sk_spell.csv')
df.head()
```

```python
import re
sen = []
for text in df['title']:
    mixed_sentences = re.findall(r'[\w\s가-힣]+[A-Za-z]+[\w\s가-힣]+', text)
    if mixed_sentences != []:
        sen.append(mixed_sentences)
```
