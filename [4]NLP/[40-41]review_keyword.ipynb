```python
# 한국어 형태소 분석

from konlpy.tag import Kkma, Komoran, Hannanum, Okt
kkma = Kkma()
komoran = Komoran()
hannanum = Hannanum()
okt = Okt()
```

```python
def keyword(data, mode, name=None): # 빈도 수를 확인해서 리뷰가 많이 등장하는 키워드 찾기
  from collections import Counter
  from konlpy.tag import Kkma, Komoran, Hannanum, Okt
  if name == None:
    data = data
  else:
    data = data[data['name']==name]
  words = ' '.join(list(data['spell']))

  # morphs_lst = komoran.morphs(words)
  pos_lst = mode.pos(words)

  # 불용어 제거
  with open('./[30]stopword_ko.txt') as f:
    stop_words = f.read().splitlines()
    f.close()

  stop_words.append('수')
  stop_words.append('곳')
  stop_words.append('것')
  stop_words.append('만')
  stop_words.append('회사')


  cleaned_words = [word for word in pos_lst if word[0] not in stop_words]

  nouns = [word for word in cleaned_words
           if word[1].startswith('N') | word[1].startswith('V')]

  count_nouns = Counter(nouns)
  count_nouns_list = count_nouns.most_common(20)

  return count_nouns_list

# 2.  TF-IDF를 이용한 회사별 & 전체 키워드
# *   회사별 키워드 분석
# *   전체 리뷰 키워드 분석

def vectorize(data, name=None): # 각 리뷰 내 키워드가 전체 문서에도 중요한 키워드인지 파악하기
  from sklearn.feature_extraction.text import TfidfVectorizer

  if name == None:
    data = data
  else:
    data = data[data['name']==name]

  data = data[data['name']== name]['spell']
  tfidf_vectorizer = TfidfVectorizer(max_features= 1000)
  tfidf_vectorizer.fit(data)
  vocab = tfidf_vectorizer.vocabulary_
  text_tfidf = tfidf_vectorizer.transform(data).toarray()

  # 가중치 0.6 이상의 단어
  weight = 0.6
  vocab_c = {v:k for k,v in vocab.items()}
  words_lst = []
  for item in text_tfidf:
    words = []
    for index, value in enumerate(item):
      if value >= weight:
        words.append(vocab_c[index])
    words_lst.append(words)

  # best_keyword
  tfidf_top = [vocab_c[text.argmax()] for text in text_tfidf]

  return pd.DataFrame(data=zip(df[df['name']==name]['spell'], words_lst, tfidf_top),
                      columns = ['리뷰', 'Above_T', 'top_keyword'])

def collocation(df, name): # 리뷰의 중요 키워드에 대해 어떤 단어가 같이 붙는지에 따라 리뷰의 형태가 달라지기에 "연어"를 확인하기
  import nltk
  from konlpy.tag import Kkma, Komoran, Hannanum, Okt
  from nltk.collocations import BigramAssocMeasures, TrigramAssocMeasures, QuadgramAssocMeasures
  from nltk.collocations import BigramCollocationFinder, TrigramCollocationFinder, QuadgramCollocationFinder

  bigram_measures = BigramAssocMeasures()
  trigram_measures = TrigramAssocMeasures()
  fourgram_measures = QuadgramAssocMeasures()

  # list의 data -> 한 문장으로
  data = df[df['name']==name]['spell']
  data = ' '.join(data)

  hannanum = Hannanum()

  word_lst = hannanum.morphs(data)

  with open('./[30]stopword_ko.txt') as f:
    stop_words = f.read().splitlines()
    f.close()

  stop_words.append('수')
  stop_words.append('곳')
  stop_words.append('것')
  stop_words.append('만')
  stop_words.append('회사')

  cleaned_words = [word for word in word_lst if word[0] not in stop_words]

  bi_ngrams = BigramCollocationFinder.from_words(cleaned_words)
  tri_ngrams = TrigramCollocationFinder.from_words(cleaned_words)
  quad_ngrams = QuadgramCollocationFinder.from_words(cleaned_words)

  bi_ngrams.apply_freq_filter(2) # 최소 2번 등장하는 n-gram 필터링
  tri_ngrams.apply_freq_filter(2) # 최소 2번 등장하는 n-gram 필터링
  quad_ngrams.apply_freq_filter(2) # 최소 2번 등장하는 n-gram 필터링

  bi_result = bi_ngrams.nbest(bigram_measures.pmi, 10)
  tri_result = tri_ngrams.nbest(trigram_measures.pmi, 10)
  quad_result = quad_ngrams.nbest(fourgram_measures.pmi, 10)

  result = pd.DataFrame({'Bi_gram': bi_result,
                       'Tri_gram': tri_result,
                       'Quad_gram': quad_result})

  return result
```

```python
# 'cjENM' 'cj제일제당' 'sk하이닉스' '네이버' '대한항공' '삼성물산' '삼성전자' '아모레퍼시픽' '카카오' '현대자동차'
# Kkma, Komoran, Hannanum, Okt
names = ['cjENM', 'cj제일제당', 'sk하이닉스', '네이버', '대한항공', '삼성물산', '삼성전자', '아모레퍼시픽', '카카오', '현대자동차']

# Kkma
keywords_kkma = pd.DataFrame(np.empty((20, 10)), columns=names)
for name in keywords_kkma.columns:
  keywords_kkma[name] = keyword(df, Kkma(),name)

keywords_komoran = pd.DataFrame(np.empty((20, 10)), columns=names)
for name in keywords_komoran.columns:
  keywords_komoran[name] = keyword(df, Komoran(),name)

keywords_hannanum = pd.DataFrame(np.empty((20, 10)), columns=names)
for name in keywords_hannanum.columns:
  keywords_hannanum[name] = keyword(df, Hannanum(),name)

keywords_Okt = pd.DataFrame(np.empty((20, 10)), columns=names)
for name in keywords_Okt.columns:
  keywords_Okt[name] = keyword(df, Okt(),name)
```
