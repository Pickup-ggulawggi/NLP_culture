```python
import os
import pandas as pd

file_lst = os.listdir('comp_40')

total = pd.DataFrame(columns=['info', 'date', 'rating', 'title','name'])

for f_name in file_lst:
    df = pd.read_csv('./comp_40/' + f_name, index_col=0)
    name = f_name.split('.csv')[0].split('_')[1]
    print(name)
    df['name'] = [name] * df.shape[0]
    
    for i in df.index:
        df['info'].iloc[i] = df['info'].iloc[i].split(' · ')[-1].split(' - ')[0]
        df['rating'].iloc[i] = df['rating'].iloc[i][-3:]

    total = pd.concat([total, df], axis=0)
```
